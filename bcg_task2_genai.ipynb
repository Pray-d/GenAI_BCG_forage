```python
import pandas as pd
import re

# Load the data into a DataFrame
data = {
    "company": ["apple", "apple", "apple", "microsoft", "microsoft", "microsoft", "tesla", "tesla", "tesla"],
    "year": [21, 22, 23, 21, 22, 23, 21, 22, 23],
    "Total Revenue": [365817, 394328, 383285, 168088, 198270, 211915, 53823, 81462, 48256],
    "Net Income": [9468, 99803, 96995, 61271, 72738, 72361, 5644, 12587, 2614],
    "Total Assets": [351002, 352755, 352583, 333779, 36484, 411976, 62131, 82338, 90591],
    "Total Revenue Growth (%)": [None, 7.79, -2.80, None, 17.96, 6.88, None, 51.35, -40.76],
    "Net Income Growth (%)": [None, 954.10, -2.81, None, 18.71, -0.52, None, 123.01, -79.23],
    "Average Assets": [352113.33, 352113.33, 352113.33, 260746.33, 260746.33, 260746.33, 78353.33, 78353.33, 78353.33],
}
df = pd.DataFrame(data)

def get_response(user_query):
    # Normalize the query to lowercase and strip extra spaces
    query = user_query.strip().lower()

    # Handle common typos by using regex replacements
    query = handle_typos(query)

    # Check if the query is asking for total revenue
    if "revenue" in query:
        company = extract_company(query)
        year = extract_year(query)
        if company and year:
            revenue = df[(df["company"] == company) & (df["year"] == year)]["Total Revenue"].values
            if revenue.size > 0:
                return f"The total revenue of {company.capitalize()} in {year} is ${revenue[0]} million."
            else:
                return "Sorry, I don't have data for that year or company."

    # Check if the query is asking for net income
    elif "income" in query:
        company = extract_company(query)
        year = extract_year(query)
        if company and year:
            net_income = df[(df["company"] == company) & (df["year"] == year)]["Net Income"].values
            if net_income.size > 0:
                return f"The net income of {company.capitalize()} in {year} is ${net_income[0]} million."
            else:
                return "Sorry, I don't have data for that year or company."

    # Check if the query is asking for total assets
    elif "assets" in query:
        company = extract_company(query)
        year = extract_year(query)
        if company and year:
            total_assets = df[(df["company"] == company) & (df["year"] == year)]["Total Assets"].values
            if total_assets.size > 0:
                return f"The total assets of {company.capitalize()} in {year} are ${total_assets[0]} million."
            else:
                return "Sorry, I don't have data for that year or company."

    # Check if the query is asking for average assets
    elif "average assets" in query:
        company = extract_company(query)
        if company:
            avg_assets = df[df["company"] == company]["Average Assets"].mean()
            return f"The average assets of {company.capitalize()} are ${avg_assets:.2f} million."

    else:
        return "Sorry, I didn't understand that question. Please try again."

def extract_company(query):
    # Define the companies using regex patterns
    companies = ["apple", "microsoft", "tesla"]
    for company in companies:
        if re.search(r'\b' + company + r'\b', query):
            return company
    return None

def extract_year(query):
    # Define regex for extracting years (21, 2021, 22, 2022, 23, 2023)
    year_pattern = r'\b(21|2021|22|2022|23|2023)\b'
    match = re.search(year_pattern, query)
    if match:
        return int(match.group(0))  # Return the matched year as integer
    return None

def handle_typos(query):
    # Common typo fixes using regex replacements
    query = re.sub(r"\brevnue\b", "revenue", query)  # fix "revenue" typo
    query = re.sub(r"\bnet income\b", "net income", query)  # fix "net income" typo
    query = re.sub(r"\bteh\b", "the", query)  # fix "the" typo
    query = re.sub(r"\bassets\b", "total assets", query)  # fix common shorthand for assets
    return query

# Command-line chatbot
if __name__ == "__main__":
    print("Welcome to the Financial Analysis Chatbot!")
    print("You can ask questions like:")
    print("'What is the total revenue of Apple in 21?'")
    print("'How much was the net income of Microsoft in 22?'")
    print("'What is the average assets of Tesla?'")

    while True:
        user_query = input("Enter your query (or type 'exit' to quit): ")
        if user_query.lower() in ["exit", "quit"]:
            print("Goodbye!")
            break
        response = get_response(user_query)
        print(response)
```
