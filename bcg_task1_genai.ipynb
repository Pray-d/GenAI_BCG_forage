```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/BCG_GENAI/kamt.csv",sep=";")
df.head(10)
```

```python

df = df.rename(columns={"company ": "company"})
df = df.rename(columns={"Cash Flow from Operating Activities.": "Cash Flow from Operating Activities"})
df.head(10)
```

```python
# Replace "microsfot-21" with "microsoft-21" in the 'company' column
df['company'] = df['company'].replace('microsfot-21', 'microsoft-21')
```

```python
import pandas as pd

df[['company', 'year']] = df['company'].str.extract(r'([a-zA-Z]+)-(\d+)')

# Reorder columns to have company and year first
cols = ['company', 'year'] + [col for col in df.columns if col not in ['company', 'year']]
df = df[cols]

# Display the modified DataFrame
print(df)
```

```python
import pandas as pd

# Assuming your DataFrame is named 'df'

# Columns to clean and convert to numeric
cols_to_clean = ['Total Revenue', 'Net Income', 'Total Assets', 'Total Liabilities', 'Cash Flow from Operating Activities']

# Remove commas, newlines, and spaces, and convert to numeric
for col in cols_to_clean:
    df[col] = (
        df[col]
        .astype(str)  # Ensure values are strings
        .str.replace(r"[,\n\s]+", "", regex=True)  # Remove commas, newlines, and spaces
        .replace("", "0")  # Replace empty strings with zero (if any)
        .astype(float)  # Convert to numeric
    )

# Display the cleaned DataFrame
print(df)
```

```python
df
```

```python
# Assuming your DataFrame is named 'df' and has 'company' and 'year' columns

# Convert the relevant columns to numeric type
numeric_cols = ['Total Revenue', 'Net Income', 'Total Assets', 'Total Liabilities', 'Cash Flow from Operating Activities']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Now, proceed with the year-over-year change calculation:
for metric in numeric_cols:
    growth_col = metric + ' Growth (%)'
    df[growth_col] = df.groupby('company')[metric].pct_change() * 100
# Explore other aggregate functions or groupings
# Example: Calculate average revenue by company
avg_revenue_by_company = df.groupby('company')['Total Revenue'].mean()
print("\nAverage Revenue by Company:")
print(avg_revenue_by_company)

# Example: Group by company and year to calculate sum of assets
assets_by_company_year = df.groupby(['company', 'year'])['Total Assets'].sum()
print("\nTotal Assets by Company and Year:")
print(assets_by_company_year)
```

```python

# Calculate the average of 'Total Assets' for each company
average_assets = df.groupby('company')['Total Assets'].mean()

# Add the average assets as a new column to the DataFrame
df['Average Assets'] = df['company'].map(average_assets)
```
